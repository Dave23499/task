```python
conn = mysql.connector.connect(user='root', password=password, host='localhost', database='bincomphptest')

if conn:
    print("Connected Successfully")
else:
    print("Connection Not Established")

select_table = """SELECT * FROM announced_pu_results"""
cursor = conn.cursor()
cursor.execute(select_table)
result = cursor.fetchall()

# Dynamically generate table headers
describe_table = "DESCRIBE announced_pu_results"
cursor.execute(describe_table)
columns = cursor.fetchall()
header = "<tr>"
for column in columns:
    header += "<td>{}</td>".format(column[0])
header += "</tr>"

# Dynamically generate table rows
rows = ""
for row in result:
    row_html = "<tr>"
    for value in row:
        row_html += "<td>{}</td>".format(value)
    row_html += "</tr>"
    rows += row_html

# Concatenate header and rows to create the final HTML table
table = "<table>{}</table>".format(header + rows)

# Generate HTML file
contents = '''
<!DOCTYPE html PUBLIC "-//W3C//DTD HTML 4.01 Transitional//EN">
<html>
<head>
<meta content="text/html; charset=ISO-8859-1" http-equiv="content-type">
<title>Python Webbrowser</title>
</head>
<body>
{}
</body>
</html>
'''.format(table)

filename = 'webbrowser.html'

def main(contents, filename):
    output = open(filename,"w")
    output.write(contents)
    output.close()

main(contents, filename)
webbrowser.open(filename)

if(conn.is_connected()):
    cursor.close()
    conn.close()
    print("MySQL connection is closed.")
```
